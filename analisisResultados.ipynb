Estandarizacion resultados

```python
#Optimizador Adam
promedioPrecisionAdam = (adam1['metrics/precision'] + adam2['metrics/precision'] + adam3['metrics/precision']) /3
promedioRecallAdam = (adam1['metrics/recall'] + adam2['metrics/recall'] + adam3['metrics/recall']) / 3
promediomAPAdam = (adam1['metrics/mAP_0.5'] + adam2['metrics/mAP_0.5'] + adam3['metrics/mAP_0.5']) / 3
promedioTrainBoxLossAdam = (adam1['train/box_loss'] + adam2['train/box_loss'] + adam3['train/box_loss']) / 3
promedioTrainObjLossAdam = (adam1['train/obj_loss'] + adam2['train/obj_loss'] + adam3['train/obj_loss']) / 3
promedioValBoxLossAdam = (adam1['val/box_loss'] + adam2['val/box_loss'] + adam3['val/box_loss']) / 3
promedioValObjLossAdam = (adam1['val/obj_loss'] + adam2['val/obj_loss'] + adam3['val/obj_loss']) / 3

#optimizador SGD
promedioPrecisionSGD = (sgd1['metrics/precision'] + sgd2['metrics/precision'] + sgd3['metrics/precision']) /3
promedioRecallSGD = (sgd1['metrics/recall'] + sgd2['metrics/recall'] + sgd3['metrics/recall']) / 3
promediomAPSGD = (sgd1['metrics/mAP_0.5'] + sgd2['metrics/mAP_0.5'] + sgd3['metrics/mAP_0.5']) / 3
promedioTrainBoxLossSGD = (sgd1['train/box_loss'] + sgd2['train/box_loss'] + sgd3['train/box_loss']) / 3
promedioTrainObjLossSGD = (sgd1['train/obj_loss'] + sgd2['train/obj_loss'] + sgd3['train/obj_loss']) / 3
promedioValBoxLossSGD = (sgd1['val/box_loss'] + sgd2['val/box_loss'] + sgd3['val/box_loss']) / 3
promedioValObjLossSGD = (sgd1['val/obj_loss'] + sgd2['val/obj_loss'] + sgd3['val/obj_loss']) / 3

#optimizador AdamW

promedioPrecisionAdamw = (adamw1['metrics/precision'] + adamw2['metrics/precision'] + adamw3['metrics/precision']) /3
promedioRecallAdamw = (adamw1['metrics/recall'] + adamw2['metrics/recall'] + adamw3['metrics/recall']) / 3
promediomAPAdamw = (adamw1['metrics/mAP_0.5'] + adamw2['metrics/mAP_0.5'] + adamw3['metrics/mAP_0.5']) / 3
promedioTrainBoxLossAdamw = (adamw1['train/box_loss'] + adamw2['train/box_loss'] + adamw3['train/box_loss']) / 3
promedioTrainObjLossAdamw = (adamw1['train/obj_loss'] + adamw2['train/obj_loss'] + adamw3['train/obj_loss']) / 3
promedioValBoxLossAdamw = (adamw1['val/box_loss'] + adamw2['val/box_loss'] + adamw3['val/box_loss']) / 3
promedioValObjLossAdamw = (adamw1['val/obj_loss'] + adamw2['val/obj_loss'] + adamw3['val/obj_loss']) / 3
```

Graficas de las metricas

```python
epocas = range(100)
#Grafica de la precision
plt.plot(epocas, promedioPrecisionAdam, label="Adam")
plt.plot(epocas, promedioPrecisionSGD, label="SGD")
plt.plot(epocas, promedioPrecisionAdamw, label="AdamW")

plt.xlabel("pocas de Entrenamiento")
plt.ylabel("Precisin")
plt.title("Comparativa de la precisin de los optimizadores")
plt.legend()
plt.show()
```

```python
#Grafica del recall
plt.plot(epocas, promedioRecallAdam, label="Adam")
plt.plot(epocas, promedioRecallSGD, label="SGD")
plt.plot(epocas, promedioRecallAdamw, label="AdamW")

plt.xlabel("pocas de Entrenamiento")
plt.ylabel("Recall")
plt.title("Comparativa del recall de los optimizadores")
plt.legend()
plt.show()
```

```python
#Grafica del mAP
#Grafica de la precision
plt.plot(epocas, promediomAPAdam, label="Adam")
plt.plot(epocas, promediomAPSGD, label="SGD")
plt.plot(epocas, promediomAPAdamw, label="AdamW")

plt.xlabel("pocas de Entrenamiento")
plt.ylabel("mAP (mean average precision)")
plt.title("Comparativa de mAP de los optimizadores")
plt.legend()
plt.show()
```

```python
#Grafica Box Loss Adam

plt.plot(epocas, promedioTrainBoxLossAdam, label="Entrenamiento")
plt.plot(epocas, promedioValBoxLossAdam, label="Validacin")

plt.xlabel("pocas de Entrenamiento")
plt.ylabel("Box Loss (Prdida de Cuadro)")
plt.title("Box Loss del optimizador Adam")
plt.legend()
plt.show()
```

```python
#Grafica Box Loss SGD

plt.plot(epocas, promedioTrainBoxLossSGD, label="Entrenamiento")
plt.plot(epocas, promedioValBoxLossSGD, label="Validacin")

plt.xlabel("pocas de Entrenamiento")
plt.ylabel("Box Loss (Prdida de Cuadro)")
plt.title("Box Loss del optimizador SGD")
plt.legend()
plt.show()
```

```python
#Grafica Box Loss ADAMW

plt.plot(epocas, promedioTrainBoxLossAdamw, label="Entrenamiento")
plt.plot(epocas, promedioValBoxLossAdamw, label="Validacin")

plt.xlabel("pocas de Entrenamiento")
plt.ylabel("Box Loss (Prdida de Cuadro)")
plt.title("Box Loss del optimizador AdamW")
plt.legend()
plt.show()
```

```python
#Graficas box loss juntas
# Crear una figura con tres subplots en una fila
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Grafica 1: Box Loss Adam
axs[0].plot(epocas, promedioTrainBoxLossAdam, label="Entrenamiento")
axs[0].plot(epocas, promedioValBoxLossAdam, label="Validacin")
axs[0].set_xlabel("pocas de Entrenamiento")
axs[0].set_ylabel("Box Loss (Prdida de Cuadro)")
axs[0].set_title("Box Loss del optimizador Adam")
axs[0].legend()


# Grafica 2: Box Loss SGD
axs[1].plot(epocas, promedioTrainBoxLossSGD, label="Entrenamiento")
axs[1].plot(epocas, promedioValBoxLossSGD, label="Validacin")
axs[1].set_xlabel("pocas de Entrenamiento")
axs[1].set_ylabel("Box Loss (Prdida de Cuadro)")
axs[1].set_title("Box Loss del optimizador SGD")
axs[1].legend()


# Grafica 3: Box Loss ADAMW
axs[2].plot(epocas, promedioTrainBoxLossAdamw, label="Entrenamiento")
axs[2].plot(epocas, promedioValBoxLossAdamw, label="Validacin")
axs[2].set_xlabel("pocas de Entrenamiento")
axs[2].set_ylabel("Box Loss (Prdida de Cuadro)")
axs[2].set_title("Box Loss del optimizador AdamW")
axs[2].legend()
# Ajustar el espacio entre subplots
#plt.tight_layout()

# Mostrar la figura con las tres grficas
plt.show()
```
