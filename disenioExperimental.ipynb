Comprobacion de los supuestos del modelo

```python
adam = pd.concat([adam1, adam2, adam3])
sgd = pd.concat([sgd1, sgd2, sgd3])
adamw = pd.concat([adamw1, adamw2, adamw3])
```

Prueba de Normalidad

```python
from scipy import stats
#Optimizadores con nombres asignados manualmente
optimizadores = {
    'adam1': adam1,
    'adam2': adam2,
    'adam3': adam3,
    'sgd1': sgd1,
    'sgd2': sgd2,
    'sgd3': sgd3,
    'adamw1': adamw1,
    'adamw2': adamw2,
    'adamw3': adamw3
}
optimizadores1 = {
    'Adam':adam,
    'SGD':sgd,
    'AdamW':adamw
}

# Prueba de normalidad (Shapiro-Wilk)
metricas = ['metrics/precision', 'metrics/recall', 'metrics/mAP_0.5', 'train/box_loss', 'val/box_loss', 'train/obj_loss', 'val/obj_loss']

# Crear un DataFrame para almacenar los resultados
resultadosNormalidad = pd.DataFrame(columns=['Optimizador', 'Métrica', 'Estadístico de prueba', 'p-valor'])

# Iterar sobre los optimizadores
for optimizador_nombre, optimizador in optimizadores1.items():
    # Iterar sobre las métricas
    for metrica in metricas:
        # Realizar la prueba de normalidad (Shapiro-Wilk)
        estadisticoPrueba, p_valor = stats.shapiro(optimizador[metrica])

        # Agregar los resultados al DataFrame
        resultadosNormalidad = pd.concat([resultadosNormalidad, pd.DataFrame.from_records([{
            'Optimizador': optimizador_nombre,
            'Métrica': metrica,
            'Estadístico de prueba': estadisticoPrueba,
            'p-valor': p_valor
        }])])
        

# Imprimir el DataFrame con los resultados
resultadosNormalidad
```

Prueba de Homocedasticidad

```python
#Prueba de homocedasticidad (Test de Levene)

# Optimizadores
optimizadores = [adam1, adam2, adam3, sgd1, sgd2, sgd3, adamw1, adamw2, adamw3]

# Métricas
metricas = ['metrics/precision', 'metrics/recall', 'metrics/mAP_0.5', 'train/box_loss', 'val/box_loss', 'train/obj_loss', 'val/obj_loss']

# Crear un DataFrame para almacenar los resultados
resultadosLevene = pd.DataFrame(columns=['Métrica', 'Estadístico de prueba', 'p-valor'])

# Iterar sobre las métricas
for metrica in metricas:
    # Obtener los datos correspondientes a la métrica
    datos_metrica = [optimizador[metrica] for optimizador in optimizadores]
    
    # Realizar el test de Levene
    estadisticoPrueba, p_valor = stats.levene(*datos_metrica)
    
    # Agregar los resultados al DataFrame
    resultadosLevene = pd.concat([resultadosLevene, pd.DataFrame.from_records([{
            'Métrica': metrica,
            'Estadístico de prueba': estadisticoPrueba,
            'p-valor': p_valor
        }])])

# Imprimir el DataFrame con los resultados

resultadosLevene
```

Comparaciones multiples de Kruskal-wallis

```python
#Prueba de Kruskal Wallis para comparaciones
from scipy.stats import kruskal
# Agrupar los datos por algoritmo y métrica de evaluación
grouped = matriz.groupby(['Algoritmo'])

# Realizar la prueba de Kruskal-Wallis para cada métrica de evaluación
for metrica in ['Precisión', 'Recall', 'mAP', 'Train Box Loss', 'Val Box Loss', 'Train Obj Loss', 'Val Obj Loss']:
    # Obtener los datos para cada algoritmo en la métrica de evaluación actual
    grupos = [group[metrica] for _, group in grouped]
    
    # Realizar la prueba de Kruskal-Wallis
    stat, p_value = kruskal(*grupos)
    
    # Imprimir los resultados
    print(f"Métrica: {metrica}")
    print(f"Estadístico de prueba: {stat}")
    print(f"Valor p: {p_value}\n")
```

Prueba Post hoc

```python
import scikit_posthocs as sp

# Realizar la prueba de Dunn para cada métrica de evaluación
for metrica in ['Precisión', 'Recall', 'mAP', 'Train Box Loss', 'Val Box Loss', 'Train Obj Loss', 'Val Obj Loss']:
    # Obtener los datos para cada algoritmo en la métrica de evaluación actual
    grupos = [group[metrica] for _, group in grouped]
    
    # Realizar la prueba de Dunn
    results = sp.posthoc_dunn(grupos, p_adjust='bonferroni')
    
    # Imprimir los resultados
    print(f"Métrica: {metrica}")
    print(results)
    print("\n")
```
